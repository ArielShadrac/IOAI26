# Pourquoi PyTorch ?

PyTorch est devenu l'un des frameworks les plus populaires pour le Deep Learning grâce à :

- sa syntaxe proche de Python ;
- sa facilité de débogage ;
- sa flexibilité pour la recherche ;
- son support des GPU ;
- son vaste écosystème.

Quelques domaines d'application :

- Vision par ordinateur
- Traitement automatique du langage naturel (NLP)
- Systèmes de recommandation
- Analyse de séries temporelles
- Intelligence artificielle générative

# Workflow général d'un projet PyTorch

Tout projet de Deep Learning suit généralement les étapes suivantes :

1. Préparer les données
2. Construire un modèle
3. Définir une fonction de perte (Loss Function)
4. Définir un optimiseur
5. Entraîner le modèle
6. Évaluer les performances
7. Sauvegarder le modèle

Dans cette formation, nous suivrons exactement cette logique.

# Vérification de l'accélération GPU
