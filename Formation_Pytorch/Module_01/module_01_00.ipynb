# Introduction aux Tensors

Un tensor (tenseur) est une structure de données multidimensionnelle utilisée par PyTorch.

Les tenseurs sont similaires aux tableaux NumPy, mais ils possèdent des fonctionnalités supplémentaires :

- Accélération GPU ;
- Calcul automatique des gradients ;
- Intégration native avec les réseaux de neurones.

Un tenseur peut être représenté comme :

- Scalaire → 0 dimension
- Vecteur → 1 dimension
- Matrice → 2 dimensions
- Tableau multidimensionnel → n dimensions

```python
import torch

print(torch.__version__)
```

# Création depuis une liste

PyTorch permet de transformer directement des listes Python en tenseurs.

```python
data = [1, 2, 3, 4]

x = torch.tensor(data)

print(x)
```

Création d'un tenseur 2D

```python
data = [
    [1,2],
    [3,4]
]

x = torch.tensor(data)

print(x)
```

# Tenseurs avec valeurs initiales

PyTorch fournit plusieurs fonctions pour créer rapidement des tenseurs.

```python
zeros = torch.zeros(3,4)

print(zeros)
```

```python
ones = torch.ones(2,3)

print(ones)
```

```python
random_tensor = torch.rand(3,3)

print(random_tensor)
```

# Attributs d'un tenseur

Un tenseur possède plusieurs propriétés importantes :

- Shape
- Type de données
- Device

```python
tensor = torch.rand(3, 4)

print(f"Forme du tensor : {tensor.shape}")
print(f"Type de données : {tensor.dtype}")
print(f"Device de stockage : {tensor.device}")
```

Ces trois attributs sont ceux que vous vérifierez le plus souvent en pratique — notamment pour déboguer un mismatch de shape entre deux tensors, ou pour vous assurer que tout est bien sur le même device avant une opération

# Manipulation des dimensions

Dans le Deep Learning, les données sont généralement représentées sous forme de tenseurs multidimensionnels.

Par exemple :

- Un vecteur peut représenter une série de valeurs numériques ;
- Une image en niveaux de gris peut être représentée par une matrice ;
- Une image couleur peut être représentée par un tenseur 3D ;

Il est donc essentiel de savoir déterminer le nombre de dimensions d'un tenseur et modifier sa forme lorsque cela est nécessaire.

## Nombre de dimensions d'un tensor

PyTorch permet de connaître le nombre de dimensions d'un tenseur grâce à l'attribut `ndim`.

Considérons la matrice suivante :

\[
\begin{bmatrix}
1 & 2 \\
3 & 4
\end{bmatrix}
\]

Cette matrice possède :

- 2 lignes ;
- 2 colonnes ;
- 2 dimensions.

Voyons comment le vérifier avec PyTorch.

```python
tensor = torch.tensor([
    [1, 2],
    [3, 4]
])

print(tensor)
print("Nombre de dimensions :", tensor.ndim)
```

Remarque :

- Un scalaire possède 0 dimension ;
- Un vecteur possède 1 dimension ;
- Une matrice possède 2 dimensions ;
- Un tenseur 3D possède 3 dimensions ;
- etc.

Cette notion est fondamentale car les réseaux de neurones attendent souvent des données dans un format précis.

## Modifier la forme d'un tensor

Dans de nombreux cas, les données doivent être réorganisées avant d'être utilisées par un modèle.

Par exemple :

- transformer un vecteur en matrice ;
- transformer une image en vecteur ;
- préparer un lot de données pour l'entraînement.

La méthode `reshape()` permet de modifier la forme d'un tenseur sans modifier les valeurs qu'il contient.

```python
x = torch.arange(12)

print("Tensor original :")
print(x)

print("Shape :", x.shape)
```

```python
x = x.reshape(3, 4)

print(x)
print("Nouvelle forme :", x.shape)
```

Le nombre total d'éléments doit rester identique.

Dans notre exemple :

- Tensor initial : 12 éléments ;
- Tensor transformé : 3 × 4 = 12 éléments.

Si le nombre d'éléments n'est pas conservé, PyTorch génère une erreur.
