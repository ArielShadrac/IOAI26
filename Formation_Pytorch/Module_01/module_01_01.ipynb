# Opérations sur les tensors

Les tenseurs supportent la plupart des opérations mathématiques classiques :

- addition ;
- soustraction ;
- multiplication ;
- division ;
- produit matriciel.

Ces opérations constituent la base des calculs effectués dans les réseaux de neurones.

## Addition

L'addition s'effectue élément par élément.


```python
import torch
```

```python
a = torch.tensor([1, 2, 3])

b = torch.tensor([4, 5, 6])

c = a + b

print(c)
```

```python
c = torch.add(a, b)

print(c)
```

```python
a.add_(b)
print(a)
```

## Multiplication élément par élément

Cette opération multiplie chaque élément du premier tenseur par l'élément correspondant du second tenseur.

```python
a = torch.tensor([1, 2, 3])

b = torch.tensor([4, 5, 6])

print(a * b)
```

Attention :

Cette opération n'est pas un produit matriciel.

Chaque élément est multiplié individuellement.

## Produit matriciel

Le produit matriciel est l'une des opérations les plus importantes en Deep Learning.

Les couches linéaires des réseaux de neurones utilisent massivement cette opération.

PyTorch permet de réaliser un produit matriciel grâce à :

- `torch.matmul()`
- l'opérateur `@`

```python
A = torch.tensor([
    [1, 2],
    [3, 4]
])

B = torch.tensor([
    [5, 6],
    [7, 8]
])

result = torch.matmul(A, B)

print(result)
```

```python
result = A @ B

print(result)
```

Les deux syntaxes produisent exactement le même résultat.

L'opérateur `@` est souvent préféré car il est plus concis et plus proche de l'écriture mathématique.

```python
x = torch.tensor([1, 2, 3, 4])

print("Somme :", torch.sum(x))
print("Moyenne :", torch.mean(x.float()))
print("Maximum :", torch.max(x))
print("Minimum :", torch.min(x))
```

Ces fonctions sont très utilisées lors :

- de l'analyse exploratoire des données ;
- du calcul de métriques ;
- de la normalisation des données.
