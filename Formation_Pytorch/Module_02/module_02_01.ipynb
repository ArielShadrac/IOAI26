# Datasets & DataLoaders

Dans un projet de Machine Learning, les données constituent l'élément central.

Avant d'entraîner un modèle, il faut être capable de :

- stocker les données ;
- accéder aux exemples individuellement ;
- transformer les données ;
- envoyer les données au modèle efficacement.

PyTorch fournit deux outils essentiels pour cette étape :

1. Dataset
2. DataLoader

## Le workflow général des données

Un pipeline classique ressemble à ceci :

Données brutes

        ↓

Dataset

        ↓

DataLoader

        ↓

Modèle PyTorch

        ↓

Entraînement


Le Dataset définit **comment accéder aux données**.

Le DataLoader définit **comment fournir ces données au modèle**.

# Création d'un Dataset personnalisé

Créons un petit dataset contenant :

- des données d'entrée X ;
- des labels associés y.

Ce principe est identique à celui utilisé pour les vrais datasets d'images, de texte ou de séries temporelles.

````python
from torch.utils.data import Dataset


class SimpleDataset(Dataset):

    def __init__(self):

        # Données d'entrée
        self.data = [
            [1,2],
            [3,4],
            [5,6],
            [7,8]
        ]

        # Labels associés
        self.labels = [
            0,
            1,
            0,
            1
        ]


    def __len__(self):

        # Retourne le nombre total d'exemples
        return len(self.data)


    def __getitem__(self, index):

        # Retourne un exemple et son label
        sample = self.data[index]
        label = self.labels[index]

        return sample, label
````

Notre Dataset contient maintenant 4 exemples :

| Donnée | Label |
|-|-|
| [1,2] | 0 |
| [3,4] | 1 |
| [5,6] | 0 |
| [7,8] | 1 |

Testons son fonctionnement.

````python
dataset = SimpleDataset()


print("Nombre d'exemples :", len(dataset))

print("Premier élément :", dataset[0])
````

Dans la pratique, les modèles PyTorch travaillent avec des tenseurs.

Nous allons donc convertir nos données en tensors.

````python
import torch


class TensorDataset(Dataset):

    def __init__(self):

        self.data = torch.tensor([
            [1,2],
            [3,4],
            [5,6],
            [7,8]
        ])

        self.labels = torch.tensor([
            0,
            1,
            0,
            1
        ])


    def __len__(self):

        return len(self.data)


    def __getitem__(self,index):

        return self.data[index], self.labels[index]
````

````python
dataset = TensorDataset()

print(dataset[0])
````

# DataLoader

Un Dataset permet d'accéder aux données une par une.

Mais pendant l'entraînement, un modèle travaille généralement avec plusieurs exemples simultanément.

Le DataLoader permet de :

- créer des batches ;
- mélanger les données ;
- charger les données progressivement.

Dataset :

1000 images


Batch size = 32


DataLoader fournit :

Batch 1 → 32 images

Batch 2 → 32 images

Batch 3 → 32 images

...

````python
from torch.utils.data import DataLoader


dataset = TensorDataset()


dataloader = DataLoader(
    dataset,
    batch_size=2
)
````

Ici :

- Dataset contient 4 exemples ;
- batch_size = 2.


Le DataLoader va donc créer :

Batch 1 → exemples 1 et 2

Batch 2 → exemples 3 et 4

````python
for batch in dataloader:

    print(batch)
````

# Shuffle

Pendant l'entraînement, il est généralement recommandé de mélanger les données.

Pourquoi ?

Parce qu'un ordre fixe peut introduire un biais dans l'apprentissage.

````python
dataloader = DataLoader(
    dataset,
    batch_size=2,
    shuffle=True
)


for batch in dataloader:

    print(batch)
````

# Pipeline complet

Nous allons maintenant construire un pipeline classique :

Dataset

↓

Transformations

↓

DataLoader

↓

Modèle

````python
from torchvision import datasets
from torchvision.transforms import ToTensor


training_data = datasets.FashionMNIST(
    root="data",
    train=True,
    download=True,
    transform=ToTensor()
)


train_loader = DataLoader(
    training_data,
    batch_size=64,
    shuffle=True
)


print(len(train_loader))
````

Le modèle recevra maintenant :

- un batch de 64 images ;
- leurs labels correspondants.

Format typique :[64, 1, 28, 28]
Cela signifie :

- 64 images ;
- 1 canal (niveau de gris) ;
- hauteur 28 pixels ;
- largeur 28 pixels.

````python
import matplotlib.pyplot as plt


images, labels = next(iter(train_loader))


plt.imshow(images[0].squeeze(), cmap="gray")

plt.title(labels[0].item())

plt.axis("off")
````
