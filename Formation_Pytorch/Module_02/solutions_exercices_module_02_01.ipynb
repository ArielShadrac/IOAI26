### Exercice 1

```python
from torch.utils.data import Dataset


class MyDataset(Dataset):

    def __init__(self):

        self.X = [
            [10,20],
            [30,40],
            [50,60]
        ]

        self.y = [
            1,
            0,
            1
        ]


    def __len__(self):
        return len(self.X)


    def __getitem__(self, index):
        return self.X[index], self.y[index]
```

```python
dataset = MyDataset()

print("Nombre d'éléments :", len(dataset))
```

```python
for i in range(len(dataset)):
    print(dataset[i])
```

### Exercice 2

```python
from torch.utils.data import DataLoader


dataloader = DataLoader(
    dataset,
    batch_size=2,
    shuffle=True
)
```

```python
for batch in dataloader:
    print(batch)
```

### Exercice 3

```python
from torchvision import datasets
from torchvision.transforms import ToTensor
from torch.utils.data import DataLoader


training_data = datasets.FashionMNIST(
    root="data",
    train=True,
    download=True,
    transform=ToTensor()
)


train_loader = DataLoader(
    training_data,
    batch_size=32,
    shuffle=True
)
```

```python
#recuperer un batch d'images et de labels
images, labels = next(iter(train_loader))
```

```python
print("Shape des images :", images.shape)

print("Shape des labels :", labels.shape)
```

```python
import matplotlib.pyplot as plt

plt.imshow(images[0].squeeze(), cmap="gray")

plt.title(f"Label : {labels[0].item()}")

plt.axis("off")
```

```python
# Vérifier les types des images et des labels
print(type(images))
print(type(labels))
```

## Que contient réellement un batch ?

Un batch est simplement un groupe d'images traitées simultanément par le réseau de neurones.

Par exemple :

torch.Size([32, 1, 28, 28])

signifie :

- 32 images
- 1 canal de couleur
- hauteur : 28 pixels
- largeur : 28 pixels

Chaque image est une matrice de nombres représentant l'intensité des pixels.

Le rôle de matplotlib est de transformer cette matrice numérique en image visible.
