### Exercice 1

````python
from torchvision import datasets
from torchvision.transforms import ToTensor

training_data = datasets.FashionMNIST(
    root="data",
    train=True,
    download=True,
    transform=ToTensor()
)

image, label = training_data[0]

print(type(image))
print(image.shape)
````

### Explications

L'image est devenue un tensor grâce à :

```python
transform=ToTensor()
```

Cette transformation convertit automatiquement l'image en tensor PyTorch.

La forme :

```python
torch.Size([1, 28, 28])
```

signifie :

- 1 canal (niveau de gris)
- 28 pixels de hauteur
- 28 pixels de largeur

## Correction Exercice 2

Nous devons créer une transformation personnalisée qui multiplie une valeur par 2.

````python
from torchvision.transforms import Lambda

transform = Lambda(
    lambda x: x * 2
)

print(transform(5))
````

## Correction Exercice 4

Créer une transformation composée à l'aide de Compose().

````python
from torchvision.transforms import Compose
from torchvision.transforms import ToTensor

transform = Compose([
    ToTensor()
])

print(transform)
````

````python
from torchvision import datasets

training_data = datasets.FashionMNIST(
    root="data",
    train=True,
    download=True,
    transform=transform
)

image, label = training_data[0]

print(type(image))
print(image.shape)
````

### Pourquoi utiliser Compose() ?

Même lorsqu'une seule transformation est utilisée, Compose() permet de préparer facilement une chaîne de transformations.

Par exemple :

```python
Compose([
    Resize((64,64)),
    ToTensor(),
    Normalize(mean=[0.5], std=[0.5])
])
```

Chaque transformation est appliquée dans l'ordre.

Cette approche rend le code plus lisible et plus facile à maintenir.
