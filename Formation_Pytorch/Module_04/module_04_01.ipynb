# Build the Neural Network

Dans les modules précédents, nous avons appris à :

- manipuler des tensors ;
- charger des données ;
- transformer les données.

Cependant, un modèle de Deep Learning doit maintenant être capable de recevoir ces données et produire une sortie.

Un réseau de neurones est une succession de couches qui transforment progressivement les données d'entrée.

Structure générale :

Input Data

↓

Neural Network

↓

Output Prediction

## Exemple

Pour une classification d'images FashionMNIST :

Entrée :

Une image 28×28 pixels


Sortie :

Une classe parmi 10 catégories :

- T-shirt
- Trouser
- Pullover
- Dress
- Coat
- ...
- Ankle Boot


Le réseau doit apprendre une fonction :

Input → Output

# torch.nn

PyTorch fournit le module :

```python
torch.nn
```

qui contient les composants nécessaires pour construire des réseaux de neurones.

Quelques éléments importants :

| Élément | Rôle |
|-|-|

````python
import torch
from torch import nn

print(torch.__version__)
````

# Les couches (*Layers*)

Une couche transforme une entrée en une sortie.

Une couche linéaire réalise l'opération :

y = xW + b


où :

- x représente l'entrée ;
- W représente les poids ;
- b représente le biais ;
- y représente la sortie.


Dans PyTorch, cette opération correspond à :

```python
nn.Linear()
```

````python
layer = nn.Linear(
    in_features=3,
    out_features=2
)

print(layer)
````

Cette couche attend :

Entrée :

3 valeurs


Sortie :

2 valeurs


Elle contient automatiquement :

- des poids ;
- un biais.

Ces paramètres seront appris pendant l'entraînement.

# Créer un modèle avec nn.Module

Dans PyTorch, tous les modèles héritent de :

```python
nn.Module
```


Un modèle doit généralement définir :

1. Les couches dans `__init__()`
2. Le chemin des données dans `forward()`

````python
class SimpleModel(nn.Module):

    def __init__(self):

        super().__init__()

        self.layer1 = nn.Linear(
            3,
            2
        )


    def forward(self, x):

        output = self.layer1(x)

        return output
````

Explication :

`__init__()`

définit l'architecture du modèle.


`forward()`

définit comment les données circulent dans le réseau.

````python
model = SimpleModel()

print(model)
````

# Forward Pass

Le forward pass correspond au passage des données à travers le réseau.

Pour l'instant, le modèle ne fait qu'une prédiction.

Il n'apprend pas encore.

````python
input_tensor = torch.tensor(
    [1.,2.,3.]
)

output = model(input_tensor)

print(output)
````

Le modèle reçoit :

```
[1,2,3]
```

et retourne :

```
[ valeur1 , valeur2 ]
```


Ces valeurs sont appelées :

**logits**

Elles représentent les scores produits par le modèle.

# Modèle de classification d'images

Nous allons maintenant construire un modèle adapté à FashionMNIST.

Les images possèdent :

28 × 28 pixels

soit :

784 valeurs.


Avant d'entrer dans une couche Linear, l'image doit être transformée en vecteur.

# Flatten

Une image possède une structure :

```
1 × 28 × 28
```


Une couche Linear attend :

```
784
```


`nn.Flatten()` réalise cette transformation.

````python
flatten = nn.Flatten()

image = torch.rand(1,28,28)

output = flatten(image)

print(output.shape)
````

# nn.Sequential

Lorsque les couches s'enchaînent simplement, PyTorch permet d'utiliser :

```python
nn.Sequential()
```

Cela permet d'écrire un modèle plus rapidement.

````python
model = nn.Sequential(

    nn.Flatten(),

    nn.Linear(784,512),

    nn.ReLU(),

    nn.Linear(512,10)

)


print(model)
````

Architecture obtenue :

Image

↓

Flatten

↓

Linear (784 → 512)

↓

ReLU

↓

Linear (512 → 10)

↓

Classes prédites

# Déplacer un modèle sur GPU

Comme les tensors, les modèles peuvent être exécutés sur GPU.

Le modèle et les données doivent être placés sur le même device.

````python
device = (
    "cuda"
    if torch.cuda.is_available()
    else "cpu"
)

print(device)
````

````python
model = model.to(device)

print(next(model.parameters()).device)
````

# Paramètres du modèle

Un réseau possède des paramètres entraînables :

- poids ;
- biais.


Ces paramètres seront ajustés pendant l'entraînement.

````python
total_params = sum(
    p.numel()
    for p in model.parameters()
)


print(total_params)
````

# Exemple complet

````python
import torch
from torch import nn


class NeuralNetwork(nn.Module):

    def __init__(self):

        super().__init__()

        self.flatten = nn.Flatten()

        self.network = nn.Sequential(
            nn.Linear(784,512),
            nn.ReLU(),
            nn.Linear(512,10)
        )


    def forward(self,x):

        x = self.flatten(x)

        logits = self.network(x)

        return logits
````

````python
model = NeuralNetwork()

print(model)
````
