## Correction Exercice 1

Une couche `nn.Linear()` réalise une transformation linéaire :

\[
y = xW^T + b
\]


Elle possède :

- un nombre d'entrées (`in_features`) ;
- un nombre de sorties (`out_features`).


Ici :

Entrée :

```
5 valeurs
```

Sortie :

```
3 valeurs
```

Donc nous devons créer :

```python
nn.Linear(5,3)
```

````python
import torch
from torch import nn


# Création de la couche
layer = nn.Linear(
    in_features=5,
    out_features=3
)


# Création d'une entrée contenant 5 valeurs
x = torch.tensor(
    [1.,2.,3.,4.,5.]
)


# Passage dans la couche
output = layer(x)


print(output)
````

## Correction Exercice 2

Nous allons créer un modèle personnalisé.

Pourquoi utiliser `nn.Module` ?

Parce qu'un réseau de neurones est un objet complexe contenant :

- des couches ;
- des paramètres ;
- une logique de calcul.


Nous allons donc définir :

1. Les couches dans `__init__()`
2. Le chemin des données dans `forward()`

````python
import torch
from torch import nn


class SimpleNetwork(nn.Module):

    def __init__(self):

        super().__init__()

        self.layer1 = nn.Linear(
            10,
            5
        )

        self.activation = nn.ReLU()

        self.layer2 = nn.Linear(
            5,
            2
        )


    def forward(self,x):

        x = self.layer1(x)

        x = self.activation(x)

        x = self.layer2(x)

        return x
````

````python
model = SimpleNetwork()

print(model)
````

````python
#Tester le modèle.
input_data = torch.rand(10)

output = model(input_data)

print(output)
````

## Correction Exercice 3

Une image FashionMNIST possède :

```
1 × 28 × 28
```

Mais une couche Linear attend un vecteur.

Nous devons donc transformer :

```
1 × 28 × 28
```

en :

```
784
```

C'est le rôle de :

```python
nn.Flatten()
```

````python
import torch
from torch import nn


class FashionModel(nn.Module):

    def __init__(self):

        super().__init__()

        self.model = nn.Sequential(

            nn.Flatten(),

            nn.Linear(784,128),

            nn.ReLU(),

            nn.Linear(128,10)

        )


    def forward(self,x):

        return self.model(x)
````

````python
#créer une instance du modèle
model = FashionModel()

print(model)
````

````python
#Tester le modèle avec une image aléatoire de taille 28x28
image = torch.rand(1,28,28)


output = model(image)


print(output.shape)
````
