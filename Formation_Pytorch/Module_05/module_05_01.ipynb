# Autograd

Autograd signifie :

Automatic Differentiation

(c'est-à-dire différentiation automatique).


Dans un réseau de neurones, les paramètres du modèle doivent être ajustés pour réduire l'erreur.

Pour cela, PyTorch doit calculer :

- les dérivées ;
- les gradients ;
- la direction dans laquelle modifier les paramètres.


Autograd réalise automatiquement ces calculs mathématiques.

## Pourquoi avons-nous besoin des gradients ?

Un modèle possède des paramètres :

- poids (weights)
- biais (bias)


Pendant l'apprentissage :

Paramètres

↓

Prédiction

↓

Erreur

↓

Calcul du gradient

↓

Mise à jour des paramètres


Les gradients indiquent comment modifier les paramètres pour améliorer le modèle.

# Exemple simple

Considérons :

$$y=x^2$$

La dérivée est :

$$\frac{dy}{dx}=2x$$

Cette dérivée indique la pente de la fonction.

Si :  
**$x = 3$**

alors :

$$\frac{dy}{dx}=6$$

Le gradient nous indique comment $y$ évolue lorsque $x$ change.

# requires_grad

Par défaut, PyTorch ne calcule pas les gradients des tensors.

Pour demander le suivi des opérations, on utilise :

```python
requires_grad=True
```

````python
import torch


x = torch.tensor(
    3.0,
    requires_grad=True
)


print(x)
````

# Computational Graph

PyTorch construit automatiquement un graphe représentant les opérations mathématiques.

Exemple :

$$y=x^2$$

Le graphe est :

```
x

↓

Square

↓

y
```


Chaque opération est enregistrée.

````python
x = torch.tensor(
    3.0,
    requires_grad=True
)


y = x ** 2


print(y)
````

Explication :

Le champ grad_fn indique que PyTorch connaît l'opération qui a produit ce tensor.

Ici PowBackward0 correspond à : $$x^2$$

# backward()

La méthode :

```python
.backward()
```

demande à PyTorch de calculer les gradients.


Elle applique automatiquement les règles de dérivation.

````python
x = torch.tensor(
    3.0,
    requires_grad=True
)


y = x ** 2


y.backward()


print(x.grad)
````

# Paramètres d'un réseau

Les poids d'un réseau sont des tensors avec :

```python
requires_grad=True
```


Ainsi PyTorch peut calculer automatiquement :

- gradient des poids ;
- gradient des biais.

````python
from torch import nn


layer = nn.Linear(
    3,
    1
)


for param in layer.parameters():

    print(param.requires_grad)
````

````python
#voir les gradients des paramètres du modèle après une passe avant et arrière
x = torch.tensor(
    [[1.,2.,3.]]
)


output = layer(x)


loss = output.sum()


loss.backward()


for name, param in layer.named_parameters():

    print(name)

    print(param.grad)
````

# torch.no_grad()

Pendant une prédiction, nous n'avons pas besoin de calculer les gradients.

Cela permet :

- d'utiliser moins de mémoire ;
- d'accélérer les calculs.

````python
with torch.no_grad():

    output = layer(x)

print(output)
````

# Workflow Autograd

Pendant l'entraînement :

1. Les données passent dans le modèle

```
input → model → prediction
```


2. On calcule l'erreur

```
prediction → loss
```


3. On calcule les gradients

```
loss.backward()
```


4. Les paramètres sont mis à jour

```
weights = weights - learning_rate × gradient
```


Autograd réalise automatiquement l'étape 3.
