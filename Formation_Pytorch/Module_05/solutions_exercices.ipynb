## Correction Exercice 1

Nous avons :

$$
y=x^2+3x
$$

avec :

$$
x=5
$$

L'objectif est :

1. Calculer y ;
2. Calculer automatiquement le gradient :

$$
\frac{dy}{dx}
$$

à l'aide de PyTorch.

````python
import torch

x = torch.tensor(
    5.0,
    requires_grad=True
)

y = x**2 + 3*x

print("Valeur de y :", y)

y.backward()

print("Gradient :", x.grad)
````

## Correction Exercice 2

Considérons :

$$
y=3x^3+2x^2+x
$$

avec :

$$
x=2
$$

Nous allons comparer :

- le résultat PyTorch ;
- le calcul mathématique manuel.

````python
import torch

x = torch.tensor(
    2.0,
    requires_grad=True
)

y = 3*x**3 + 2*x**2 + x

print("y =", y)

y.backward()

print("Gradient =", x.grad)
````

## Correction Exercice 3

Nous allons créer une couche :

```python
nn.Linear(4,2)
```

Cette couche possède :

- 4 entrées ;
- 2 sorties.

Elle contient automatiquement :

- une matrice de poids ;
- un biais.

L'objectif est d'observer les gradients calculés par Autograd.

````python
import torch
from torch import nn

layer = nn.Linear(
    4,
    2
)

x = torch.tensor(
    [[1.,2.,3.,4.]]
)

output = layer(x)

print("Sortie :")
print(output)
````

````python
loss = output.sum()

print("Loss :")
print(loss)
````

````python
loss.backward()
````

````python
for name, param in layer.named_parameters():

    print(f"\n{name}")

    print(param.grad)
````

# Points clés

Autograd est le moteur de calcul automatique des gradients de PyTorch.

Pour qu'un tensor soit suivi :

```python
requires_grad=True
```

Pour calculer les gradients :

```python
loss.backward()
```

Pour consulter un gradient :

```python
tensor.grad
```

Pour éviter le calcul des gradients :

```python
with torch.no_grad():
```

Ces gradients seront utilisés dans le prochain module pour entraîner un réseau de neurones à l'aide d'un optimiseur.
