```python
#Exercice 1 
import torch
from torch import nn

loss_fn = nn.CrossEntropyLoss()

predictions = torch.tensor(
    [[1.0, 3.0, 0.5]]
)

target = torch.tensor([1])

loss = loss_fn(
    predictions,
    target
)

print(loss)
```

```python
#Exercice 3
import torch
from torch import nn

model = nn.Linear(
    3,
    2
)

X = torch.rand(
    1,
    3
)

y = torch.tensor([1])

loss_fn = nn.CrossEntropyLoss()

optimizer = torch.optim.SGD(
    model.parameters(),
    lr=0.01
)
```

```python
pred = model(X)

print(pred)
```

```python
loss = loss_fn(
    pred,
    y
)

print(loss)
```

```python
loss.backward()
```

```python
for name, param in model.named_parameters():

    print(name)

    print(param.grad)
```

```python
optimizer.step()
```

```python
optimizer.zero_grad()
```
