# Save and Load Model

L'entraînement d'un modèle peut prendre :

- quelques secondes ;
- plusieurs minutes ;
- plusieurs heures ;
- plusieurs jours.

Il serait inefficace de réentraîner le modèle à chaque utilisation.

PyTorch permet donc de sauvegarder les paramètres appris afin de les réutiliser ultérieurement.

Exemples :

- Sauvegarder un modèle après l'entraînement ;
- Charger un modèle pour faire des prédictions ;
- Continuer un entraînement interrompu ;
- Déployer un modèle dans une application.

# state_dict()

Chaque modèle possède :

- des poids (weights)
- des biais (bias)

Ces paramètres sont stockés dans :

```python
model.state_dict()
```

Le `state_dict` est un dictionnaire Python contenant tous les paramètres entraînables du modèle.

````python
import torch
from torch import nn

model = nn.Linear(3,2)

print(model.state_dict())
````

Le `state_dict` contient uniquement les paramètres.

Il ne contient pas :

- le code du modèle ;
- les fonctions Python ;
- les données d'entraînement.

C'est pour cette raison qu'il est généralement recommandé de sauvegarder le `state_dict`.

# torch.save()

Pour sauvegarder les paramètres d'un modèle :

```python
torch.save(
    model.state_dict(),
    "model.pth"
)
```

Le fichier `.pth` contiendra tous les poids du modèle.

````python
model = nn.Linear(
    3,
    2
)

torch.save(
    model.state_dict(),
    "model.pth"
)

print("Model saved.")
````

Après exécution, un fichier apparaît dans votre dossier de travail :

```text
model.pth
```

Ce fichier contient les paramètres du modèle.

# Important

Avant de charger les poids :

il faut recréer exactement la même architecture.

PyTorch doit savoir où placer les paramètres chargés.

````python
model = nn.Linear(
    3,
    2
)
````

Si le modèle original est :

```python
nn.Linear(3,2)
```

le modèle chargé doit être :

```python
nn.Linear(3,2)
```

Sinon les dimensions des paramètres ne correspondront pas.

# torch.load()

Pour restaurer les paramètres :

```python
model.load_state_dict(
    torch.load("model.pth")
)
```

````python
model = nn.Linear(
    3,
    2
)

model.load_state_dict(
    torch.load("model.pth")
)

print("Model loaded.")
````

# model.eval()

Après chargement :

```python
model.eval()
```

place le modèle en mode évaluation.

C'est une étape importante avant d'effectuer des prédictions.

Certaines couches se comportent différemment :

- Dropout
- BatchNorm

Pendant l'entraînement :

```python
model.train()
```

Pendant l'inférence :

```python
model.eval()
```

Même si notre modèle actuel ne contient pas ces couches, il est recommandé de toujours utiliser `model.eval()`.

# Inference

Une fois chargé :

le modèle peut être utilisé pour prédire de nouvelles données.

````python
with torch.no_grad():

    x = torch.tensor(
        [1.,2.,3.]
    )

    prediction = model(x)

    print(prediction)
````

Nous utilisons :

```python
torch.no_grad()
```

car :

- aucun entraînement n'est effectué ;
- aucun gradient n'est nécessaire.

Cela réduit l'utilisation mémoire.

````python
#Exemple complet de sauvegarde et de chargement d'un modèle PyTorch

import torch
from torch import nn

# Création du modèle

model = nn.Linear(
    3,
    2
)

# Sauvegarde

torch.save(
    model.state_dict(),
    "model.pth"
)

# Recréation du modèle

loaded_model = nn.Linear(
    3,
    2
)

# Chargement

loaded_model.load_state_dict(
    torch.load("model.pth")
)

# Mode évaluation

loaded_model.eval()

print("Model ready.")
````
